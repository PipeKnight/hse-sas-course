```python
categorical
```

```python
numeric = ["day", "day_year", "PRICE_REGULAR", "PRICE_AFTER_DISC"]
```

```python
def my_smape(A, F):
    A = np.expm1(A)
    F = np.expm1(F)
    return 100/len(A) * np.sum(2 * np.abs(F - A) / (np.abs(A) + np.abs(F)))

smape = make_scorer(my_smape, greater_is_better=False)
```

```python
xgb_flags = xgb.XGBClassifier(n_estimators=120, learning_rate=0.22, max_depth=10, 
                              reg_lambda=1, eval_metric="aucpr")
```

```python
column_transformer = ColumnTransformer(
    [
        ("ohe", OneHotEncoder(handle_unknown="ignore"), categorical),
        ("scaling", StandardScaler(), numeric),
    ]
)

pipeline = Pipeline(
    steps=[
        ("ohe_and_scaling", column_transformer),
        (
            "GBRegressor",
            xgb_flags,
        ),
    ]
)
```

> Сделаем метрику accuracy и запустим кросс-валидацию, чтобы посмотреть качество классификатора:

```python
my_accuracy_score = make_scorer(f1_score, greater_is_better=True)
```

```python
model = pipeline
print(cross_val_score(model, X_train, y_train, cv=7, scoring=my_accuracy_score).sum() / 7)
```

```python
model.fit(X_train, y_train)
y_temp = model.predict(X_test)
accuracy_score(y_test, y_temp)
```

```python
args = {'eval_metric': 'mae', 'eta': 0.2, 'n_estimators': 120, 'max_depth': 8}
xgb_cool = xgb.XGBRegressor(**args)

args = {'eval_metric': 'mae', 'eta': 0.25, 'n_estimators': 150, 'max_depth': 6}
xgb_cool2 = xgb.XGBRegressor(**args)

features = np.array(X.columns)

features = features[
    ~np.isin(
        features,
        ["flag_zero", "demand", "flag_much", "id", "50%", "mean_demand"],
    )
]

categorical = features[
    ~np.isin(
        features,
        ["demand", "50%", "mean_demand", "PRICE_REGULAR", "PRICE_AFTER_DISC"],
    )
]

numeric = ["day", "day_year", "PRICE_REGULAR", "PRICE_AFTER_DISC"]

forest2 = RandomForestRegressor(n_estimators=12, max_depth=6, n_jobs=-1)

forest3 = RandomForestRegressor(n_estimators=10, max_depth=5, n_jobs=-1)

forest4 = RandomForestRegressor(n_estimators=14, max_depth=8, n_jobs=-1)

from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import RidgeCV

linreg2 = Ridge(alpha = 5)

estimators = [
    ('lr', forest2),
    ('svr', xgb_cool2),
    ('svr2', xgb_cool),
    ('ugabuga', forest3),
    ('linrega', forest4),
    ('linreg', linreg2),
]

reg = StackingRegressor(
    estimators=estimators,
    final_estimator=RidgeCV(cv=None, scoring="neg_root_mean_squared_error", alphas=np.logspace(-2, 3, 100))
)

column_transformer = ColumnTransformer(
    [
        ("ohe", OneHotEncoder(handle_unknown="ignore"), categorical),
        ("scaling", StandardScaler(), numeric),
    ]
)

pipeline = Pipeline(
    steps=[
        ("ohe_and_scaling", column_transformer),
        (
            "GBRegressor",
            reg,
        ),
    ]
)
```
